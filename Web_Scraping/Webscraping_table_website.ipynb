```python
url = "https://ticker.finology.in/"

# getting request
r = requests.get(url)
print(r)
soup = BeautifulSoup(r.text,"lxml")
```

```python
table = soup.find("table",class_ = "table table-sm table-hover screenertable")
print(table)
```

```python
#obtain the headers

#th ---> table headers
#tr ---> table row
#td ---> table data


headers = table.find_all("th")
print(headers)
```

```python
title = []

for i in headers:
    names = i.text
    title.append(names)
    
print(title)
```

```python
#making dataframe with those headers

df = pd.DataFrame(columns = title)
print(df)
```

```python
# creating a loop that will first find all the "tr" tags., inside that creating another loop., finding all "td" tags
```

```python
rows = soup.find_all("tr")
print(rows)
```

```python
#index 0 is table header, we already got them
#so start from index 1

for i in rows[1:]:
    print(i.text)
    
```

```python
#obtaining in differant way

#finding for table data tags

for i in rows[1:]:
    data = table.find_all("td")
    print(data)
```

```python
#obtaining in differant way

#printing all the td data

rows = table.find_all("tr")

for i in rows[1:]:
    data = i.find_all("td")
    #print(data)
    row = [j.text for j in data]
    print(row)
    
    
```

```python
#obtaining in differant way

#printing all the td data

rows = table.find_all("tr")

for i in rows[1:]:
    data = i.find_all("td")
    #print(data)
    row = [j.text for j in data]
    print(row)
    l = len(df)
    df.loc[l] = row
    
print(df)
    
```
