```python
# using Beutiful Soup

import requests
from bs4 import BeautifulSoup

url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"

# getting request
r = requests.get(url)
print(r)
soup = BeautifulSoup(r.text,"lxml")
```

```python
#---------- find() -----------------------------------

#finding everything inside div tag
print(soup.find("div"))
print("\n")
print(soup.div)  #this also does the same

print("\n")
price = (soup.find("h4",{"class":"price float-end card-title pull-right"}))
print(price.string)

print("\n")
desc = (soup.find("p",{"class":"description"}))
print(desc.string)
print('\n')
#or
desc = (soup.find("p",class_="description"))
print(desc.string)

#the main disadvantage with find is ., it will print whatever it finds first. if we want 2nd or 4th things it may not find well
```

```python
#------------------ find all() ---------------------------------------

'''find all help us to find all data., related to that perticular tag'''
print("\n")
price = (soup.find_all("h4",class_="price float-end card-title pull-right"))
print(price)
print(len(price))

for i in price:
    #print(i)            #prints entire class attribute string
    #print(i.text)     #prints entire class attribute string
    print(i.string)   #prints only price

print(" ")
#finding galaxy description
desc = soup.find_all("p",class_="description")
print(desc[3])
```

```python
#------------------ find all() with Regex---------------------------------------
import re
#--> printing all h4,a,p tags
data = soup.find_all(['h4','a','p'])
print(data)
```

```python
data = soup.find_all(string = "Galaxy Tab")
print(data)
#find all help us to find exact string
```

```python
#using regular expression

#regular expression used to find string related to given pattern
data = soup.find_all(string = re.compile("Galaxy Tab"))
print(data)
```

```python
#using regular expression
data = soup.find_all(string = re.compile("Galaxy"))
print(data)
print(len(data))
```

```python
#finding how many times the word lenovo appears in the entire page

data = soup.find_all(string = re.compile("Lenovo"))
print(data)
```
